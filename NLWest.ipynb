```python
import pybaseball as pyball
from pybaseball import team_pitching
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = 2023
dodgers = pyball.schedule_and_record(year, 'LAD')
dbacks = pyball.schedule_and_record(year, 'ARI')
rockies = pyball.schedule_and_record(year, 'COL')
padres = pyball.schedule_and_record(year, 'SDP')
giants = pyball.schedule_and_record(year, 'SFG')

dodgers
```

```python
dodgers['Wins'] = np.where(dodgers['W/L']=='W', 1, (np.where(dodgers['W/L']=='W-wo', 1, 0))).cumsum()
dodgers['Losses'] = np.where(dodgers['W/L']=='L', 1, (np.where(dodgers['W/L']=='L-wo', 1, 0))).cumsum()
dodgers['Win_Percentage'] = dodgers['Wins'] / (dodgers['Wins'] + dodgers['Losses'])

dbacks['Wins'] = np.where(dbacks['W/L']=='W', 1, (np.where(dbacks['W/L']=='W-wo', 1, 0))).cumsum()
dbacks['Losses'] = np.where(dbacks['W/L']=='L', 1, (np.where(dbacks['W/L']=='L-wo', 1, 0))).cumsum()
dbacks['Win_Percentage'] = dbacks['Wins'] / (dbacks['Wins'] + dbacks['Losses'])

rockies['Wins'] = np.where(rockies['W/L']=='W', 1, (np.where(rockies['W/L']=='W-wo', 1, 0))).cumsum()
rockies['Losses'] = np.where(rockies['W/L']=='L', 1, (np.where(rockies['W/L']=='L-wo', 1, 0))).cumsum()
rockies['Win_Percentage'] = rockies['Wins'] / (rockies['Wins'] + rockies['Losses'])

padres['Wins'] = np.where(padres['W/L']=='W', 1, (np.where(padres['W/L']=='W-wo', 1, 0))).cumsum()
padres['Losses'] = np.where(padres['W/L']=='L', 1, (np.where(padres['W/L']=='L-wo', 1, 0))).cumsum()
padres['Win_Percentage'] = padres['Wins'] / (padres['Wins'] + padres['Losses'])

giants['Wins'] = np.where(giants['W/L']=='W', 1, (np.where(giants['W/L']=='W-wo', 1, 0))).cumsum()
giants['Losses'] = np.where(giants['W/L']=='L', 1, (np.where(giants['W/L']=='L-wo', 1, 0))).cumsum()
giants['Win_Percentage'] = giants['Wins'] / (giants['Wins'] + giants['Losses'])
#gets wins to date, losses to date, and win percentage to date for each team
```

```python
#now plotting a graph for total wins for each team throughout the 2023 season
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(dodgers['Wins'], label='Dodgers', c='blue')
plt.plot(dbacks['Wins'], label='dbacks', c='red')
plt.plot(rockies['Wins'], label='rockies', c='green')
plt.plot(padres['Wins'], label='padres', c='orange')
plt.plot(giants['Wins'], label='giants', c='brown')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Wins')

plt.legend(loc='lower right')

plt.title(f"NL west Wins Comparison ({year})")
```

```python
#Now we plot for the win percentage of each team

plt.rcParams["figure.figsize"] = (8,6)

plt.plot(dodgers['Win_Percentage'], label='Dodgers', c='blue')
plt.plot(dbacks['Win_Percentage'], label='Dbacks', c='red')
plt.plot(rockies['Win_Percentage'], label='Rockies', c='green')
plt.plot(padres['Win_Percentage'], label='rockies', c='orange')
plt.plot(giants['Win_Percentage'], label='SF Giants', c='brown')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Win Percentage')

plt.legend(loc='lower right')

plt.title(f"AL East Win Percentage Comparison ({year})")
```

```python
#total runs allowed by each team 
dodgers['Total_Runs'] = dodgers['R'].cumsum()
dodgers['Total_RA'] = dodgers['RA'].cumsum()

dbacks['Total_Runs'] = dbacks['R'].cumsum()
dbacks['Total_RA'] = dbacks['RA'].cumsum()

rockies['Total_Runs'] = rockies['R'].cumsum()
rockies['Total_RA'] = rockies['RA'].cumsum()

padres['Total_Runs'] = padres['R'].cumsum()
padres['Total_RA'] = padres['RA'].cumsum()

giants['Total_Runs'] = giants['R'].cumsum()
giants['Total_RA'] = giants['RA'].cumsum()

#Now we graph it
```

```python
plt.rcParams["figure.figsize"] = (16,8)

plt.subplot(1,2,1)
plt.plot(dodgers['Total_Runs'], label='Dodgers', c='blue')
plt.plot(dbacks['Total_Runs'], label='DBacks', c='red')
plt.plot(rockies['Total_Runs'], label='Rockies', c='green')
plt.plot(padres['Total_Runs'], label='Padres', c='orange')
plt.plot(giants['Total_Runs'], label='Giants', c='brown')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs')

plt.title(f"AL East Runs Comparison ({year})")
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(dodgers['Total_RA'], label='Dodgers', c='navy')
plt.plot(dbacks['Total_RA'], label='DBacks', c='red')
plt.plot(rockies['Total_RA'], label='Rockies', c='green')
plt.plot(padres['Total_RA'], label='Padres', c='orange')
plt.plot(giants['Total_RA'], label='Giants', c='brown')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs Allowed')

plt.title(f"AL East Runs Allowed Comparison ({year})")
plt.legend(loc='lower right')

# A comparison of runs scored to runs allowed
```

```python
team_pitching_stats = team_pitching(2023, end_season=2023, league='NL', ind=0)
team_pitching_stats
#team pitching stats for every team in National League
```

```python
dodgers_p = team_pitching_stats[team_pitching_stats['Team'] == 'LAD']
dbacks_p = team_pitching_stats[team_pitching_stats['Team'] == 'ARI']
rockies_p = team_pitching_stats[team_pitching_stats['Team'] == 'COL']
padres_p = team_pitching_stats[team_pitching_stats['Team'] == 'SDP']
giants_p = team_pitching_stats[team_pitching_stats['Team'] == 'SFG']

era_2023 = {'Team': ['Dodgers', 'DBacks', 'Rockies', 'Padres', 'SF Giants'],
        'ERA': [dodgers_p['ERA'].sum(), dbacks_p['ERA'].sum(), rockies_p['ERA'].sum(), padres_p['ERA'].sum(), giants_p['ERA'].sum()]
        }

df_era = pd.DataFrame(era_2023)

print(df_era.sort_values(by='ERA', ascending=False))
```
